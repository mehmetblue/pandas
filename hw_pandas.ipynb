```python
# Q4. Find out the types values of the columns and save the result as a dataframe.

types = pd.DataFrame(df.dtypes, columns=['types'])
types
```

```python
# Q5. Find out the non-null counts of the columns and save the result as a dataframe.

notnulls = pd.DataFrame(df.notnull().sum(), columns=['not nulls'])
notnulls
```

```python
# Q6. Find out the null counts of the columns and save the result as a dataframe.

isnulls = pd.DataFrame(df.isnull().sum(), columns=['nulls'])
isnulls
```

```python
# Q7. Find out the unique counts of the columns and save the result as a dataframe.

uniques = pd.DataFrame(df.nunique(), columns=['unuques'])
uniques
```

```python
# Q8. Merge the dataframes you created in questions 4-5-6-7.

columns = pd.DataFrame(df.columns, columns=['columns'])
columns.reset_index(drop=True, inplace=True)
types.reset_index(drop=True, inplace=True)
notnulls.reset_index(drop=True, inplace=True)
isnulls.reset_index(drop=True, inplace=True)
uniques.reset_index(drop=True, inplace=True)

list = [columns, types, notnulls, isnulls, uniques]

df_concated = pd.concat(list, join='outer', axis=1)

df_concated
```

```python
# Q9. Lowercase all column names.

df.columns = map(str.lower, df.columns)
df.columns
```

```python
# Q10. Change all the No values to NoRain and all the Yes values to Rain in raintoday and raintomorrow columns.

df.raintoday.replace(['No', 'Yes'], ['NoRain', 'Rain'], inplace=True)
df.raintomorrow.replace(['No', 'Yes'], ['NoRain', 'Rain'], inplace=True)
df.loc[:,'raintoday':'raintomorrow']
```

```python
# Q11. Change the data type of "date" (object) column to datetime64 and reformat the date as DD/MM/YYYY.

df.date.dtype
```

```python
# change date columns' datatype to datetime64[ns] to able to change the date format
df['date'] = df['date'].astype('datetime64[ns]')
print(df.date.dtype)
print(df.date)

print(20*'*')

# chnange the date format from 'yyyy-mm-dd' to 'dd/mm/yyyy'
df['date'] = df['date'].dt.strftime('%d/%m/%y')
print(df.date.dtype)
print(df.date)

print(20*'*')

# again, change date columns' datatype to datetime64[ns]
df['date'] = df['date'].astype('datetime64[ns]')
print(df.date.dtype)
print(df.date)
```

```python
# Q12. Create a new column called "difference", calculate the difference between maxtemp and mintemp columns for each row, and store the value in this new column.

pd.DataFrame(df['maxtemp']-df['mintemp'], columns=['difference'])
```

```python
# Q13. Remove the evaporation and sunshine columns from the dataset permanently.

df.drop(['evaporation', 'sunshine'], axis=1)
```

```python
# Q14. Find out the most rainy day for each city.

# group the locations and find the max values in the column according to the unique location names
pd.DataFrame(df.loc[df.groupby('location').rainfall.idxmax()], columns=['location', 'date','rainfall'])
```

```python
# Q15. Filter out all the data for the city 'Albury' and then sort according to maxtemp column.

df[df['location'] == 'Albury'].sort_values(by=['maxtemp'], ascending=False)
```

```python
# Q16. Find out the NaN counts for each column.

pd.DataFrame(df.isnull().sum(axis=0), columns=['NaN_s'])
```

```python
# Q17. Remove the rows with NaN values in "windgustdir" column from the dataframe permanently.

df.dropna(subset=['windgustdir'], axis=0, inplace=True)
df
```

```python
# Q18. Create a new dataframe, use "Location" column as the index of the dataframe, display the min, max, and median values of "evaporation" and "sunshine" columns in this dataframe.
n1df = df.copy()

n2df = n1df.groupby("location")[["evaporation","sunshine"]]
n2df.min()
```

```python
n2df.max()
```

```python
n2df.median()
```

```python
# Q19. Find out the hottest day of "Perth". Example output: Timestamp('2015-01-05 00:00:00')

perth = df[df['location'] == 'Perth'].sort_values(by=['maxtemp'], ascending=False)

perth.iloc[0][['location', 'date', 'maxtemp']]
```
